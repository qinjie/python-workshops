```python
import requests 
from bs4 import BeautifulSoup
```

```python
# URL to scrape
url = 'https://exchange-rates.org/converter/SGD/EUR/100'
```

```python
# retrieve the HTML content - server rendered
resp = requests.get(url)
```

```python
html = resp.text
print(html)
```

```python
soup = BeautifulSoup(html)
```

```python
from_amount = soup.find('span', { 'id': 'ctl00_M_lblFromAmount' })
from_currency = soup.find('span', { 'id': 'ctl00_M_lblFromIsoCode'})
```

```python
print(from_amount.text, from_currency.text)
```

```python
to_amount = soup.find('span', { 'id': 'ctl00_M_lblToAmount'})
to_currency = soup.find('span', { 'id': 'ctl00_M_lblToIsoCode'})
```

```python
print(to_amount.text, to_currency.text)
```

```python
print(f'{from_amount.text} {from_currency.text} = {to_amount.text} {to_currency.text}')
```
